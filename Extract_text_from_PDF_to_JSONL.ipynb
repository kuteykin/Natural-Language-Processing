# Extract text from PDF using *pdfplumber*

```python
import pdfplumber
```

```python
pdf = pdfplumber.open("Input/The-lord-of-the-rings.eng.pdf")
```

```python
pdf.metadata
```

```python
page_10 = pdf.pages[10]
```

```python
page_10.extract_text()
```

```python
page_10.to_image(resolution=150)
```

```python
page_10.extract_words()
```

```python
page_10.crop((40,50, 370, 580)).to_image(resolution=150)
```
